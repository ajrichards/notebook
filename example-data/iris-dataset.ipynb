## Iris dataset

There are many other example datasets available.  Some of the easiest ones to find are:

   * http://www.statsmodels.org/devel/datasets/index.html#loading-data-as-pandas-objects
   * http://scikit-learn.org/stable/datasets/index.html#loading-from-external-datasets
   * http://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
## make imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set(style="ticks", color_codes=True)

## load the iris data
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
```

```python
g = sns.pairplot(iris, hue="species",palette="husl")
```
